```python
#Importing essentials
from __future__ import division
import numpy as np
import os
import preprocess
```

```python
from time import gmtime, strftime
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
import pandas as pd
```

```python
datetime = strftime("%Y-%m-%d %H:%M:%S", gmtime())
```

```python
#Importing essentials from directory Utah_project(Subcellular Localization)
curr_dir = os.getcwd()
seqdir = curr_dir + "/seqs/"
seqfiles = os.listdir(seqdir)
props_file = "aa_propierties.csv"
add_props = True
seq_len = 200
```

```python
#data preprocessing
dataset = preprocess.DataSet(seqdir, props_file, add_props, seq_len)
```

```python
#Sequence length append
lengths = []
for subclass in dataset.raw_seqs:
    for seq in subclass:
        lengths.append(len(seq))
```

```python
#Plotting the data with aa length on x-axis 
#seed setting
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "distributions")))
sns.distplot(lengths, axlabel= "Length (aa)", color="r")
plt.xlim([0,2000])

plt.show()
```
